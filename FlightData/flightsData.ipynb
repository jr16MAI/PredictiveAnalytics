```python
import pandas as pd
import os

folder_path = '/Users/judith.rethmann/Downloads/FlightsData'

dataframes = []

def convert_to_time(x):
    if pd.isna(x) or x == 0:
        return pd.to_datetime('00:00', format='%H:%M').time()
    x = f'{int(x):04d}'
    if x == '2400':
        return pd.to_datetime('23:59', format='%H:%M').time()
    return pd.to_datetime(f'{x[:2]}:{x[2:]}', format='%H:%M').time()

for filename in os.listdir(folder_path):
    if filename.startswith('T_ONTIME_REPORTING') and filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path, parse_dates=['FL_DATE'])
        
        # Ensure FL_DATE is in datetime format
        df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])
        
        dataframes.append(df)

# Find common columns across all dataframes
common_columns = set(dataframes[0].columns)
for df in dataframes[1:]:
    common_columns.intersection_update(df.columns)

common_columns = list(common_columns)
df_flight_data = pd.concat([df[common_columns] for df in dataframes], ignore_index=True)

# Convert DEP_TIME to datetime.time format after concatenation
df_flight_data['DEP_TIME'] = df_flight_data['DEP_TIME'].fillna(0).apply(convert_to_time)

# Filter the data for flights from San Francisco (SFO) by United Airlines (UA)
df_flights_SFO_UA = df_flight_data[
    (df_flight_data['ORIGIN_CITY_NAME'] == 'San Francisco, CA') & 
    (df_flight_data['OP_UNIQUE_CARRIER'] == 'UA') & 
    (df_flight_data['CANCELLED'] == 0) & 
    (df_flight_data['DIVERTED'] == 0)
]

columns_to_keep = [
    'ORIGIN_CITY_NAME', 'ORIGIN_AIRPORT_ID', 'OP_UNIQUE_CARRIER', 'OP_CARRIER_FL_NUM', 'TAIL_NUM',
    'FL_DATE', 'YEAR', 'MONTH', 'DAY_OF_MONTH', 'DAY_OF_WEEK', 'DEST_CITY_NAME',
    'DEST_AIRPORT_ID', 'AIR_TIME', 'DEP_TIME', 'DEP_DELAY', 'DEP_DEL15'
]

df_flights_SFO_UA = df_flights_SFO_UA[columns_to_keep]

# Save the filtered dataframe to a CSV file
df_flights_SFO_UA.to_csv('/Users/judith.rethmann/Documents/MBS/PredictiveAnalytics/FlightData/flights_SFO_UA.csv', index=False)
```
